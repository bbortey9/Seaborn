```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#To view the plots
%matplotlib inline
```

```python
#importing data
Auto_Mpg_data=pd.read_csv('/content/drive/MyDrive/Projects/auto_data.csv')
```

```python
#Viewing heads of data
Auto_Mpg_data.head(5)
```

```python
#write a user defined function for origin
#1-USA, 2-Europe, 3-Asia
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Asia'
#Use apply function
Auto_Mpg_data['origin']=Auto_Mpg_data['origin'].apply(origin)
```

```python
#View 1st 30 data points after applying user defind functions
Auto_Mpg_data.head(30)
```

```python
#Draw the pair plot using sns for mpg, weight,origin and with hue origin,set the size to 4.
#note:hue is variable in data set to map plot aspects to different colors
sns.pairplot(Auto_Mpg_data[['mpg','weight','origin']], hue='origin')
```
